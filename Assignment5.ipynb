클래스와 모듈은 모두 하나의 이름공간을 의미 합니다.


적절한 이름을 관리함으로서 보다 효율적인 코딩을 가능하게 하는 도구라는 것이 공통점입니다.

차이점으로 클래스는 개채로 인스턴스화 시킬 수 있지만, 표준 모듈은 개체로 인스턴스화 될 수 없다는 것입니다 

표준 모듈의 데이터 복사본은 하나만 존재할 수 있으므로 프로그램의 한 부분에서 표준 모듈의 공용 변수를 변경하면, 

프로그램의 다른 부분에서도 해당 변수를 읽을 때 동일한 값을 얻습니다. 반면 개체 데이터는 인스턴스화 된 각 개체마다 별도로 존재합니다. 


뿐만 아니라 표준모듈과 달리 클래스는 인터페이스를 구현할 수 있습니다.

```python
class UniverseClass:
    def level(self):
        print 'easy'
        
class Script(UniverseClass):
    def level(self):
        print 'Hell'
        
class DataStructure(UniverseClass):
    def level(self):
        print 'Difficult'
        
class Java(UniverseClass):
    pass

for each in (Script(), DataStructure(), Java()):
    each.level()
```

# 4번문제

\_\_init__() 메소드 함수는 생성자를 뜻합니다. 객체가 생성되면서 자동으로 초기화를 시켜줘야 하는 부분을 담당합니다. 

이 메소드 init()에 있는 코드 내용은 매개변수로 받는 리스트 길이를 받고, 그만큼 내용을 새로운 리스트에 추가시키는 내용입니다.

\_\_str()__ 메소드 함수는 리스트에 있는 내용을 보기 좋게 출력하는 메소드 입니다.

본래 리스트는 [ ] 안에 있도록 출력이 되는데, \_\_str\_\_ 코드 내용은 { } 안에 있도록 출력
을 하는 내용으로 구현되어 있습니다.

eliminae_duplicate 함수의 코드 내용은 중복된 내용은 추가시키지 않도록 구현되어 있습니다. for문으로 검사 후 s에 있지 않는 내용이라면 추가 시킵니다.
그렇지 않으면 계속 반복문을 진행하도록 구현돼 있습니다.

if \_\_name__ == "_\_\main\_\_": 구문은 있어도 없어도 되는 구문이지만, 메인함수임을 표시해주는 역할입니다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, B):         #요구사항 1
        s = []
        for e in self:           #s 리스트에 self 리스트 변수가 없으면 리스트에 추가한다.
            if e not in s:
                s.append(e)
        for e in B:              #s 리스트에 B 리스트 변수가 없으면 리스트에 추가한다.
            if e not in s:
                s.append(e)
                
        result = "MySet: {"      # 결과 출력
        for e in s:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    def __and__(self, B):       # 요구사항 2
        s = []
        for e in B:             #이중 반복문으로 B 리스트에 Self 변수가 있는 것만 리스트에 추가한다
            if e in self:
                s.append(e)
                
        result = "MySet: {"     #결과 출력
        for e in s:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    def __sub__(self,B):         #요구사항 3
        s = []
        for e in self:           #s 리스트에 self리스트 변수들을 추가한다
            s.append(e)
        for e in B:              #s 리스트에 B리스트 변수가 있으면 리스트에서 제거한다
            if e in self:
                s.remove(e)
                
        result = "MySet: {"      #결과 출력
        for e in s:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
        
     
```

# 6번문제

bool 이나 len을 구현하지 않은 경우, 기본적으로 사용자 정의 클래스의 객체는
참된 값이라고 간주합니다 bool(x)는 x.__bool__()을 호출한 결과를 이용합니다.

__bool__()이 구현되어 있지 않으면 파이썬은 x.__len__()을 호출하며 이 특별 메서드가 0을 반환하면 bool()은 False를, 그렇지 않으면 True를 반환합니다.

in 역시 마찬가지의 이유입니다.


```python
def sum_Solution(x):
    x=str(x) 
    num=0
    for i in range(len(x)):
        num+=int(x[i])
    return num

sum_Solution(2**1000)
```
